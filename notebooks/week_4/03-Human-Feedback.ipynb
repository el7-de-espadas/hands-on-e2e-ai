### Import Dependencies

```python
from langsmith import Client
```

### Attach feedback to a specific trace in langsmith

```python
client = Client()
```

```python
RUN_ID="019f7831-9680-7cc2-829d-9ed9c3dd4308"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    feedback_source_type="api",
    value="This is a test feedback"
)
```
